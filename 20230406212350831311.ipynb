```python
# Read content from an actual file with utf-8 encoding and handle errors
with open(r'C:\Users\Hp\Downloads\New folder (6)\35-0.txt', 'r', encoding='utf-8', errors='ignore') as file:
    file_content = file.read()

# Split the text into words
rawfile = file_content.split()

# Shuffle the characters within each word
for word_index in range(len(rawfile)):
    thisword = rawfile[word_index]
    idx = np.random.permutation(len(thisword))
    rawfile[word_index] = ''.join([thisword[i] for i in idx])

# Join the shuffled words back into a single string
shuffled_text = ' '.join(rawfile)

# Save the shuffled text to a file
with open(r'C:\Users\Hp\Downloads\New folder (6)\shuffled_text.txt', 'w', encoding='utf-8') as file:
    file.write(shuffled_text)

print("Shuffled text saved successfully.")
```
